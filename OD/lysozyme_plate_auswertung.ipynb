```python
def my_fitting(x, y, fitdict, key = None):
    
    '''
    adds a fitted curve to the plot.
    returns a new fitdict, if key is set.
    '''
    
    OD0 = y[0]
    a = np.min(y[-30:])-0.05
    m = 1/1.160 #mg/ml
    
    def make_param(OD, t, a, OD0):
        
        return (np.log((OD0 -a) / (OD - a))) / t
    
    def make_result(OD, t, m, OD0):

        return (m * (OD0 - OD)) / t
    
    
    def fitfunc(param, KM, VMax):
        '''
        OD of 0 is hardcoded in the function and slope m is also 
        hardcoded
        '''
       
        return VMax - (KM * param)
    
    def int_mm(OD, KM, VMax, a, m):
        '''
        returns the time, when is at OD is set
        '''
        
        S = m * (OD - a)
        S0 = m * (OD0 - a)

        temp1 = (KM * np.log(S / S0)) + (S - S0)
        
        return - (temp1 / VMax)
            
    #plt.plot(make_param(y[1:6], x[1:6], a, OD0), make_result(y[1:6], x[1:6], m, OD0), "bo")
    #plt.xlim(xmin = 0)
    #plt.ylim(ymin = 0)
    #the fitting
    popt, pcov = curve_fit(fitfunc, make_param(y[1:6], x[1:6], a, OD0), make_result(y[1:6], x[1:6], m, OD0), p0=(1., 0.02))
    
    KMfit, KMfiterr = popt[0], pcov[0,0]
    VMaxfit, VMaxfiterr = popt[1], pcov[1,1]
    
    OD_fit = np.linspace(np.max(y) + 0.1, np.min(y) - 0.01, 100)    
    t_fit = int_mm(OD_fit, KMfit, VMaxfit, a, m)
    t_mod = int_mm(y, KMfit, VMaxfit, a, m)
    
    #chi squared
    redchisq = redchisqg(y, t_mod, deg = 3)
    plt.plot(t_fit, OD_fit, "r--", lw = 3)
    #plt.text(0.9 * x[-1], np.mean(y), "$V_{max}" + " = $ {Vmax:5.3f} +- {Vmaxerr:5.3f}".format(Vmax = VMaxfit,
                                                                                               #Vmaxerr = VMaxfiterr))
    print popt
    print pcov
    #except:
    #    redchisq = None
    #    Vmax = None
    #    Vmaxerr = None
        
    #sort out misfitted
    if VMaxfiterr > 0.5 * VMaxfit:
        redchisq = None
        VMaxfit = None
        VMaxfiterr = None
    if key != None:
        fitdict[key] = [VMaxfit, VMaxfiterr, redchisq]
        return fitdict
```
