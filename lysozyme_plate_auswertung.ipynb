```python
#date can have a _2 for second assay on a date
DATE = "2014_08_26"
FOLDER = "Nils2_" + DATE + "/"
resultsfile = open(FOLDER + "26082014-002.asc" , "r" )
resultsfile = resultsfile.readlines() # make a list of it

#limits for plots
UPPERLIMIT = 1
LOWERLIMIT = 0.01

#the axis file has the format x-axis, then a linebreak, then the y axis. comments can be just under the second line,
#they won't be read

variationdict = {}
whatiswheredict = {}
axisfile = open(FOLDER + "axis.txt", "r")

xaxis = {}
yaxis = {}

for line in axisfile:
    line = line.strip()
    columns = line.split()
    if columns[0] == "x":
        xaxis[columns[1]] = columns[2:]
        whatiswheredict[columns[1]] = "xaxis"
        variationdict[columns[1]] = "x"
        
    elif columns[0] == "y":
        yaxis[columns[1]] = columns[2:]
        whatiswheredict[columns[1]] = "yaxis"
        variationdict[columns[1]] = columns[0]
        
    elif columns[0] == "all":
        whatiswheredict[columns[1]] = columns[2]
        variationdict[columns[1]] = columns[0]

for key in xaxis:
    xaxis[key] = np.array(xaxis[key])
    coordx = np.array((np.arange(1,len(xaxis[key]) + 1)), dtype=str)
for key in yaxis:
    yaxis[key] = np.array(yaxis[key])




axisfile.close()


coordlist = []
coordy = ["A", "B", "C","D","E","F","G","H"]

equivalentaxes = {}
for j in range(len(coordx)):
    equivalentaxes[coordx[j]] = []
    for i in range(len(coordx)):
        another = True
        for key in xaxis:
            another = another & (xaxis[key][i] == xaxis[key][j])
        if another: 
            equivalentaxes[coordx[j]].append(coordx[i])
for j in range(len(coordy)):
    equivalentaxes[coordy[j]] = []
    for i in range(len(coordy)):
        another = True
        for key in yaxis:
            another = another & (yaxis[key][i] == yaxis[key][j])
        if another: 
            equivalentaxes[coordy[j]].append(coordy[i])
            

totalredundanciesdict = {}

for entry in coordx:
    listx = equivalentaxes[entry]
    for yentry in coordy:
        listy = equivalentaxes[yentry]
        dictlist = []
        for x in listx:
            for y in listy:
                dictlist.append(y + x)
        totalredundanciesdict[yentry + entry] = dictlist
            
for j in range(len(coordy)):
    for i in range (len(coordx)):
        coordlist.append(coordy[j] + coordx[i])       
        
for key in whatiswheredict:
    if whatiswheredict[key] == "xaxis":
        xelongtemp = np.tile(xaxis[key], len(coordy))
        tempdict = {}
        for i in range(len(coordlist)):
            tempdict[coordlist[i]] = xelongtemp[i]
        whatiswheredict[key] = tempdict
        
    elif whatiswheredict[key] == "yaxis":
        yelongtemp = np.repeat(yaxis[key], len(coordx))
        tempdict = {}
        for i in range(len(coordlist)):
            tempdict[coordlist[i]] = yelongtemp[i]
        whatiswheredict[key] = tempdict
    else:
        whatiswheredict[key] = dict((coord, whatiswheredict[key]) for coord in coordlist )



        
    
        
coordtoinputdict = {}        
for coord in coordlist:
    if (variationdict["substratedilution"] == "x") | (variationdict["substratedilution"] == "y"):
        coordtoinputdict[coord] = (whatiswheredict["lysozyme"][coord] + "_dil._at " + whatiswheredict["dilution"][coord] 
                                   + "at " + whatiswheredict["temperature"][coord] + "°C" + "with_substr._dil. " + 
                                   whatiswheredict["substratedilution"][coord])
    else:    
        coordtoinputdict[coord] = (whatiswheredict["lysozyme"][coord] + "_dil._at " + whatiswheredict["dilution"][coord] 
                                   + "at_" + whatiswheredict["temperature"][coord] + "°C")
    
        


resultsarray = np.empty((len(resultsfile) - 1, (len(coordx) * len(coordy)) + 2), dtype=np.object)

tempinfile = False
for i in range(len(resultsfile)-1):
    line = resultsfile[i].strip()
    columns = line.split()
    columns = np.array(columns)
    resultsarray[i] = columns
    tempinfile = tempinfile | (columns[3] == "°C")
    


  
datadict = {}



datadict = dict( (coordlist[i],np.array(resultsarray[:,i+2], dtype=float)) for i in  range(len(coordlist)))
datadict["time"] = np.array(resultsarray[:,0], dtype=float)
datadict["time"] = datadict["time"] / 60.
if tempinfile:
    datadict["measuretemperature"] = np.array(resultsarray[:,2], dtype=float)
```

```python
def medium_of_wells(welltuple, datadict):
    '''
    returns a two arrays, the medium and the std array
    '''
    
    for well in welltuple:
        if well == welltuple[0]:
            calcarray = np.array([datadict[well]])
        else:
            calcarray = np.append(calcarray,  np.array([datadict[well]]), axis = 0)
    medarray = np.mean(calcarray, axis = 0)
    errarray = np.std(calcarray, axis = 0)
    return medarray, errarray
```

```python
def make_plots_on_axis(datadict, equivalentaxesdict, coordx, coordy):
    """
makes plots on all the axis, if two axis are exactly the same, they are combined.
    """
    skiplistx = []
    for x in coordx:
        datalist = []
        alreadylistx = [] #with coordinates
        if x not in skiplistx:
            skiplistx = skiplistx + equivalentaxesdict[x]
            skiplisttotal = []
            for y in coordy:
                if (y + x) not in skiplisttotal:
                    plotlist = []
                    for eq in totalredundanciesdict[y + x]:
                        skiplisttotal.append(eq)
                        plotlist.append(eq)
                    medarray, erraray = medium_of_wells(plotlist, datadict)
                    datalist.append((medarray, erraray, coordtoinputdict[plotlist[0]]))
            make_plots_from_datalist(datalist, datadict["time"], "{this}_axis".format(this = x))
        
    skiplisty = []
    for y in coordy:
        datalist = []
        alreadylisty = [] #with coordinates
        if y not in skiplisty:
            skiplisty = skiplisty + equivalentaxesdict[y]
            skiplisttotal = []
            for x in coordx:
                if (y + x) not in skiplisttotal:
                    plotlist = []
                    for eq in totalredundanciesdict[y + x]:
                        skiplisttotal.append(eq)
                        plotlist.append(eq)
                    medarray, erraray = medium_of_wells(plotlist, datadict)
                    datalist.append((medarray, erraray, coordtoinputdict[plotlist[0]]))
            make_plots_from_datalist(datalist, datadict["time"], "{this}_axis".format(this = y))
```

```python
def get_coord_with_properties(whatiswheredict, kind, value):
    '''
    takes a kind like "temperature" and a value and returns an array of coordinates, that have this property
    '''
    retlist = []
    for key in whatiswheredict[kind]:
        if (whatiswheredict[kind][key] == value):
            retlist.append(key)
    return np.array(retlist)
```
