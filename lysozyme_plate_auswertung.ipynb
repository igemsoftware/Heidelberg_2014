```python
import numpy as np
import matplotlib
matplotlib.use("GTK")
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def fitfunc(x, A, lam, C):
    return A * np.exp( - lam * x) + C

def make_plots(thedict, plotlisttuple = None, fit = False):
    """
Creates histograms of the distribution of all turn sequences, 
always a histogram for the distance and for the angles is
produced. These are saved in the folder called path.

Plotslisttuple always needs a list and a string of the file's name
    """
    dilutions = []
    temps= []
    keys= []
    slopes= []
    slopeerrs= []
    redchisqs= []
    lys = []
    
    if plotlisttuple == None:
        x = thedict["time"]
        for key in thedict:
            if key != "time":
            
                y = thedict[key]  
                plt.plot(x, y, "bo", ms=10)
                plt.title("OD600 of {well}".format(well = key))
                plt.xlabel("time in [min]")
                plt.ylabel("OD")
                if max(y) > UPPERLIMIT:
                    plt.ylim(ymin = LOWERLIMIT)
                else:
                    plt.ylim((LOWERLIMIT, UPPERLIMIT))

                if fit:
                    try:
                        #the fitting
                        popt, pcov = curve_fit(fitfunc, x, y)

                        x_fit = np.linspace(0., x[-1] + 10, 100)
                        y_fit = fitfunc(x_fit, *popt)
                        y_mod = fitfunc(x, *popt)
                        #chi²
                        redchisq = redchisqg(y, y_mod, deg = 3)
                        slope = popt[1]
                        try:
                            slopeerr = pcov[1,1]
                        except:
                            slopeerr = None
                        plt.plot(x_fit, y_fit, "r--", lw = 3,   
                                 label="fit at red chi² = " + "{chisq:5.2f}".format(chisq = redchisq))
                        plt.text(0.9 * x[-1], np.mean(y), "slope : {steig}".format(steig = slope))
                        
                        plt.legend()
                    except:
                        redchisq = None
                        slope = None
                        slopeerr = None

                temp = key[0:4]
                dilution = key[9:14]

                dilutions.append(dilution)
                temps.append(temp)
                keys.append(key)
                slopes.append(slope)
                slopeerrs.append(slopeerr)
                redchisqs.append(redchisq)
                lys.append(key[14] == "w")
                plt.savefig(FOLDER + "automated_plots/plot_of_{well}.png".format(well = key), dpi = 200)
                plt.close()
        return np.array(dilutions), np.array(temps), np.array(keys), np.array(slopes),\
        np.array(slopeerrs), np.array(redchisqs), np.array(lys)
            
            
    else:
        plotlist = plotlisttuple[0]
        maxcount = False
        
        x = thedict["time"]
        count = 0
        for key in plotlist:
            y = thedict[key]
            if max(y) > UPPERLIMIT:
                maxcount = True
                
            if count < 7:
                plt.plot(x, y, label = "{well}".format(well = key), lw = 3.)
            elif count < 14:
                plt.plot(x, y, "x",  label = "{well}".format(well = key), ms=10.)
            else:
                plt.plot(x, y, "o",  label = "{well}".format(well = key), ms=10.)
            count += 1
            
            
            if fit:
                try:
                    #the fitting
                    popt, pcov = curve_fit(fitfunc, x, y)

                    x_fit = np.linspace(0., x[-1] + 10, 100)
                    y_fit = fitfunc(x_fit, *popt)
                    y_mod = fitfunc(x, *popt)
                    #chi²
                    redchisq = redchisqg(y, y_mod, deg = 3)
                    slope = popt[1]
                    try:
                        slopeerr = pcov[1,1]
                    except:
                        slopeerr = None
                    plt.plot(x_fit, y_fit, lw = 3,   
                             label="fit at red chi² = " + "{chisq:5.2f}".format(chisq = redchisq))
                    
                except:
                    redchisq = None
                    slope = None
                    slopeerr = None
        if maxcount:
            plt.ylim(ymin = LOWERLIMIT)
        else:
            plt.ylim((LOWERLIMIT, UPPERLIMIT))
            
        plt.title("OD600 of {well}".format(well = plotlisttuple[1]))
        plt.xlabel("time in [min]")
        plt.ylabel("OD")        
        plt.legend()
        plt.savefig( FOLDER + "plot_of_{well}.png".format(well = plotlisttuple[1]), dpi=200)
        plt.close()    
        if fit:
            temp = key[0:4]
            dilution = key[9:14]

            dilutions.append(dilution)
            temps.append(temp)
            keys.append(key)
            slopes.append(slope)
            slopeerrs.append(slopeerr)
            redchisqs.append(redchisq)
            lys.append(key[14] == "w")
            return np.array(dilutions), np.array(temps), np.array(keys), np.array(slopes),\
        np.array(slopeerrs), np.array(redchisqs), np.array(lys)
        
def redchisqg(ydata,ymod,deg=2,sd=None):
    """  
Returns the reduced chi-square error statistic for an arbitrary model,   
chisq/nu, where nu is the number of degrees of freedom. If individual   
standard deviations (array sd) are supplied, then the chi-square error   
statistic is computed as the sum of squared errors divided by the standard   
deviations. See http://en.wikipedia.org/wiki/Goodness_of_fit for reference.  

ydata,ymod,sd assumed to be Numpy arrays. deg integer.  

Usage:  
>>> chisq=redchisqg(ydata,ymod,n,sd)  
where  
ydata : data  
ymod : model evaluated at the same x points as ydata  
n : number of free parameters in the model  
sd : uncertainties in ydata  

Rodrigo Nemmen  
http://goo.gl/8S1Oo  
   """  
      # Chi-square statistic  
    if sd==None:  
       chisq=np.sum((ydata-ymod)**2)  
    else:  
       chisq=np.sum( ((ydata-ymod)/sd)**2 )  
         
    # Number of degrees of freedom assuming 2 free parameters  
    nu=len(ydata)-1-deg  
    
    return chisq/nu    
```

```python
FOLDER = "Nils_2014_08_19/"
resultsfile = open(FOLDER + "19082014-003.asc" , "r" )
resultsfile = resultsfile.readlines() # make a list of it
#limits for plots
UPPERLIMIT = 0.8
LOWERLIMIT = 0.0

#the axis file has the format x-axis, then a linebreak, then the y axis. comments can be just under the second line,
#they won't be read
axisfile = open(FOLDER + "axis.txt", "r")

line = axisfile.readline()
line = line.strip()
xaxis = line.split()

line = axisfile.readline()
line = line.strip()
yaxis = line.split()
masterlist = []

for j in range(0,8):
    for i in range (len(xaxis)):
        masterlist.append(xaxis[i] + "degC_" + yaxis[j])

'''

#if temp is vertical
for j in range(0,8):
    for i in range (len(xaxis)):
        masterlist.append(yaxis[j] + "degC_" + xaxis[i])
'''    

masterlist = np.array(masterlist)


resultsarray = np.empty((len(resultsfile) - 1, (len(xaxis) * len(yaxis)) + 2), dtype=np.object)


for i in range(len(resultsfile)-1):
    line = resultsfile[i].strip()
    columns = line.split()
    columns = np.array(columns)
    resultsarray[i] = columns
    


  
```

```python
datadict = {}



datadict = dict( (masterlist[i],np.array(resultsarray[:,i+2], dtype=float)) for i in  range(len(masterlist)))
datadict["time"] = np.array(resultsarray[:,0], dtype=float)
datadict["time"] = datadict["time"] / 60.
```

```python
(dilution, temp, key, slope, slopeerr, redchisq, lys) = make_plots(datadict, fit=True) 
```

```python
def make_plots_on_axis(thedict, tobeplotted, namelist = None):
    """
takes either a list or a single entry in a list. Makes a plot with all the curves that contain the string of the list. 
It plots all data that contains a certain pattern
    """

    for i in range(len(tobeplotted)):
        wlist = []
        for entry in masterlist:
            if tobeplotted[i] in entry:
                wlist.append(entry)

        wlist.sort()
        if namelist == None:
            make_plots(thedict, plotlisttuple=(wlist, "mult" + tobeplotted[i]), fit = False)
        else:
            make_plots(thedict, plotlisttuple=(wlist, "mult" + namelist[i]), fit = False)
```

```python
make_plots_on_axis(datadict, xaxis)
make_plots_on_axis(datadict, yaxis)
```
